```python
def problem_2():
    # Initialize the list of Fibonacci numbers with the first two numbers
    fibonacci = [1, 2]
    i = 1

    # Generate Fibonacci numbers less than or equal to 4 * 10 ** 6
    while fibonacci[i] + fibonacci[i - 1] <= 4 * 10 ** 6:
        fibonacci.append(fibonacci[i] + fibonacci[i - 1])
        i += 1

    # Calculate the sum of even numbers in the Fibonacci list
    sum = 0
    for x in fibonacci:
        if x % 2 == 0:
            sum += x
    return sum

print(problem_2())
```
