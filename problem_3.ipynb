```python
import math

def problem_3(num):

    # Initialize a list to store prime numbers
    prime = []

    # Find all prime numbers up to the square root of the given number
    for i in range(2, int(math.sqrt(num)) + 1):
        for j in range(1, i+1):
            if i % j == 0 and j != 1 and j != i:
                break
            elif j > int(math.sqrt(i)):
                prime.append(i)
                break

    # Initialize a list to store prime factors of the given number
    prime_factors = []

    # Find all prime factors of the given number
    for i in prime:
        if num % i == 0:
            prime_factors.append(i)

    # Return the largest prime factor of the given number
    return max(prime_factors)

print(problem_3(600851475143))
```
