```python
def problem_4():
    # Initialize a list to store palindromes
    palindromes = []

    # Iterate over all 3-digit numbers in reverse order
    for i in range(999, 110, -1):
        for j in range(999, 110, -1):
            # Check if the product of the two numbers is a palindrome
            if str(j * i)[::-1] == str(j * i):
                # If it is a palindrome, add it to the list of palindromes
                palindromes.append(j * i)

    # Return the largest palindrome from the list of palindromes
    return max(palindromes)

print(problem_4())
```
